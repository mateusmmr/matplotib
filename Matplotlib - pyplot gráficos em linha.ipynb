# Matplotlib(pyplot) - Gráficos em linha

#### Notebook no qual treino plotagem de gráficos em linha utilizando a Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Fazendo com que gráficos sejam plotados na mesma linha:

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x ** 2
```

```python
y
```

#### Gráfico funcional:

```python
plt.plot(x,y,'g--')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Título')
```

#### Plotando vários gráficos de uma vez

```python
plt.subplot(1, 2, 1)
plt.plot(x,y,'g--')
plt.subplot(1, 2, 2)
plt.plot(y,x,'r*-')
```

O problema do subplot é que não dá muito controle sobre os gráficos:

#### Alternativa:

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # (Espaçamento à direita, espaçamento de baixo, largura, altura)
axes2 = fig.add_axes([0.2,0.5,0.3,0.2])

axes1.plot(x,y)
axes1.set_xlabel('Eixo x')
axes1.set_title('Título')
axes2.plot(y,x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, 'y--')
ax.plot(x, x**4, 'r-.')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Titulo')
```

#### Criando vários gráficos:

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
ax[0].plot(x,y, 'b')
ax[0].set_title('Titulo a')

ax[1].plot(x,y, 'g')
ax[1].set_title('Titulo b')
fig
```

#### Utilizando o tight_layout:

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
```

Nota-se que os gráficos se sobrepoem.

#### Com o tight_layout

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout()
```

#### Figuras:

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(8,6)) #ajustando tamanho da figura
axes.plot(x, y, 'r-', label = 'x^2') # label = legenda
axes.plot(y, x, 'g--', label = 'x**1/2')
axes.set_title('Titulo')
axes.legend(loc=0) #loc determina a posição da legenda - 0 é automático
```

#### Algumas opções de edição do gráfico:

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x, x**2, color='black')
ax.plot(x, x**3, color= '#D419CE', linewidth=10, alpha=0.5, linestyle=':') # cor pode ser dada em RGB e espessura da linha pode ser alterada, alpha está relacionado a transparência da linha, linestyle ao desenho da linha
ax.set_xlim([0,2]) #Alterando métrica do gráfico
ax.set_ylim([0,10])
```
